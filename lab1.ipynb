# correlation

```python
corr = df.corr()
corr 
```

# positive correlations: 
- international reputation and overall -- this makes sense as the better a player is, the more renowned they generally are.
- overall and potential -- this also makes sense as the greater a player is, the more potential they usually present; as they refine their craft.
- aggresion and penalties -- this is a very logical correlation, but I am surprised it is not stronger.

# negative correlations:
- age and potential -- this correlation proves sensical, as the older you are, the less potential you have. This is likely because you have less time left within your career.
- gk kicking and short passing -- this correlaton also makes sense, as short passing and goal kicking are very different kicks of significantly different length. Players may short pass for a reason, as their goal kick is potentially not as strong.


As far as using specific columns for inputs and targets, I would use columns such as Age and overall as inputs, as they seem to be on the causation side of things. I would put values such as international reputation and salary as targets, as they seem to be more effects of other factors (dependent variables).
